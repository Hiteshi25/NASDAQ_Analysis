```python
import yfinance as yf
import pandas as pd
import time

# 👉 load symbols
df = pd.read_csv(r"C:\Users\hites\OneDrive\Documents\MSC DATA SCIENCE\PROJECTS\NASDAQ ANALYSIS\NYSE DATA\NYSE.csv")

df.columns = df.columns.str.strip().str.lower()
symbols = df["symbol"].dropna().tolist()

# 👉 loop
for symbol in symbols:
    print(f"Downloading {symbol}...")

    try:
        data = yf.download(symbol, period="20y", interval="1d")

        if data.empty:
            print(f"No data: {symbol}")
            continue

        data.reset_index(inplace=True)
        data["symbol"] = symbol

        data = data[[
            "symbol", "Date", "Open", "Low", "High", "Close", "Volume"
        ]]

        data.columns = [
            "symbol", "date", "open_price",
            "low_price", "high_price",
            "close_price", "volume"
        ]

        data.to_csv(f"{symbol}_20y.csv", index=False)

        time.sleep(1)

    except Exception as e:
        print(f"Error: {symbol}")
```
