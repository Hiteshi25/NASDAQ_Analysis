**DATA COLLECTION TO DB**

```python
import pandas as pd
import yfinance as yf
from sqlalchemy import create_engine
from datetime import datetime, timedelta
from urllib.parse import quote_plus

# 🔹 DB connection
password = quote_plus("Hi@240504")
engine = create_engine(f"postgresql://postgres:{password}@127.0.0.1:5432/postgres")

# =========================
# 🔹 GET SYMBOLS FROM DB
# =========================
symbols_df = pd.read_sql("SELECT symbol FROM sec_master", engine)
symbols = symbols_df['symbol'].tolist()

# =========================
# 🔹 DATE RANGE (20 years)
# =========================
end_date = datetime.today()
start_date = end_date - timedelta(days=365*20)

# =========================
# 🔹 FETCH + INSERT TRADES
# =========================
for i, sym in enumerate(symbols[:]):   # ⚠️ test first
    try:
        # 🔥 FETCH DATA (YOU MISSED THIS)
        data = yf.download(sym, start=start_date, end=end_date, auto_adjust=True, progress=False)

        if data is None or data.empty:
            print(f"⚠️ Skipped: {sym}")
            continue

        # 🔥 FIX MULTIINDEX ISSUE
        data.columns = data.columns.get_level_values(0)

        data.reset_index(inplace=True)
        data['symbol'] = sym

        data.rename(columns={
            'Date': 'trade_date',
            'Open': 'open_price',
            'High': 'high_price',
            'Low': 'low_price',
            'Close': 'close_price',
            'Volume': 'volume'
        }, inplace=True)

        data = data[['symbol','trade_date','open_price','high_price','low_price','close_price','volume']]

        # 🔥 INSERT INSIDE LOOP
        data.to_sql('trades', engine, if_exists='append', index=False)

        print(f"✅ Done: {sym} ({i})")

    except Exception as e:
        print(f"❌ Error: {sym} → {e}")

print("🎉 DONE")
```

**SuperTrend Strategy using Weekly Data (2022 onwards)**

```python
import pandas as pd
import pandas_ta as ta
import numpy as np
from sqlalchemy import create_engine
from urllib.parse import quote_plus

# ====================================
# PARAMETERS (CHANGE HERE ONLY)
# ====================================

# RSI
RSI_LENGTH = 14

# SUPERTREND ENTRY
ENTRY_ST_LENGTH = 7
ENTRY_ST_FACTOR = 1

# SUPERTREND EXIT
EXIT_ST_LENGTH = 7
EXIT_ST_FACTOR = 1

# ====================================
# DB CONNECTION
# ====================================
password = quote_plus("Hi@240504")

engine = create_engine(
    f"postgresql://postgres:{password}@127.0.0.1:5432/postgres"
)

# ====================================
# LOAD WEEKLY DATA
# ====================================
df = pd.read_sql("""

SELECT *
FROM sec_trade_weekly
WHERE week_start >= '2022-01-01'
ORDER BY symbol, week_start

""", engine)

print("✅ Weekly data loaded")

# ====================================
# FINAL RESULT LIST
# ====================================
final_trades = []

# ====================================
# PROCESS SYMBOL-WISE
# ====================================
for symbol in df['symbol'].unique():

    try:

        stock = df[df['symbol'] == symbol].copy()

        stock.reset_index(drop=True, inplace=True)

        # skip small data
        if len(stock) < 20:
            continue

        # =================================
        # RSI
        # =================================
        stock['rsi'] = ta.rsi(
            stock['weekly_close'],
            length=RSI_LENGTH
        )

        # =================================
        # ENTRY SUPERTREND
        # =================================
        st_entry = ta.supertrend(
            high=stock['weekly_high'],
            low=stock['weekly_low'],
            close=stock['weekly_close'],
            length=ENTRY_ST_LENGTH,
            multiplier=ENTRY_ST_FACTOR
        )

        st_entry.columns = [
            f"ENTRY_{col}"
            for col in st_entry.columns
        ]

        stock = pd.concat([stock, st_entry], axis=1)

        # =================================
        # EXIT SUPERTREND
        # =================================
        st_exit = ta.supertrend(
            high=stock['weekly_high'],
            low=stock['weekly_low'],
            close=stock['weekly_close'],
            length=EXIT_ST_LENGTH,
            multiplier=EXIT_ST_FACTOR
        )

        st_exit.columns = [
            f"EXIT_{col}"
            for col in st_exit.columns
        ]

        stock = pd.concat([stock, st_exit], axis=1)

        # =================================
        # FIND ENTRY SUPERTREND COLUMN
        # =================================
        entry_col = [
            col for col in stock.columns
            if 'ENTRY_SUPERT_' in col
            and 'SUPERTd_' not in col
            and 'SUPERTl_' not in col
            and 'SUPERTs_' not in col
        ][0]

        # =================================
        # FIND EXIT SUPERTREND COLUMN
        # =================================
        exit_col = [
            col for col in stock.columns
            if 'EXIT_SUPERT_' in col
            and 'SUPERTd_' not in col
            and 'SUPERTl_' not in col
            and 'SUPERTs_' not in col
        ][0]

        # =================================
        # ENTRY SIGNAL
        # =================================
        stock['entry_signal'] = np.where(
            stock['weekly_close'] > stock[entry_col],
            'BUY',
            'SELL'
        )

        # =================================
        # EXIT SIGNAL
        # =================================
        stock['exit_signal'] = np.where(
            stock['weekly_close'] < stock[exit_col],
            'SELL',
            'BUY'
        )

        # =================================
        # ENTRY / EXIT LOGIC
        # =================================
        in_position = False
        entry_data = {}

        for i in range(1, len(stock)):

            current_entry = stock.loc[i, 'entry_signal']
            prev_entry = stock.loc[i - 1, 'entry_signal']

            current_exit = stock.loc[i, 'exit_signal']
            prev_exit = stock.loc[i - 1, 'exit_signal']

            # =============================
            # BUY ENTRY
            # =============================
            if (
                not in_position
                and current_entry == 'BUY'
                and prev_entry == 'SELL'
            ):

                entry_data = {

                    'symbol':
                        symbol,

                    'entry_date':
                        stock.loc[i, 'week_start'],

                    'entry_price':
                        stock.loc[i, 'weekly_close'],

                    'entry_rsi':
                        round(stock.loc[i, 'rsi'], 2),

                    'entry_rsi_past_1_week':
                        round(stock.loc[i - 1, 'rsi'], 2)
                        if i >= 1 else None,

                    'entry_rsi_past_2_week':
                        round(stock.loc[i - 2, 'rsi'], 2)
                        if i >= 2 else None
                }

                in_position = True

            # =============================
            # SELL EXIT
            # =============================
            elif (
                in_position
                and current_exit == 'SELL'
                and prev_exit == 'BUY'
            ):

                exit_price = stock.loc[i, 'weekly_close']

                gain_pct = (
                    (
                        exit_price
                        - entry_data['entry_price']
                    )
                    / entry_data['entry_price']
                ) * 100

                trade = {

                    'symbol':
                        entry_data['symbol'],

                    'entry_date':
                        entry_data['entry_date'],

                    'entry_price':
                        round(
                            entry_data['entry_price'],
                            2
                        ),

                    'entry_rsi':
                        entry_data['entry_rsi'],

                    'exit_date':
                        stock.loc[i, 'week_start'],

                    'exit_price':
                        round(exit_price, 2),

                    'exit_rsi':
                        round(stock.loc[i, 'rsi'], 2),

                    'gain_pct':
                        round(gain_pct, 2),

                    'rsi_current':
                        round(stock.loc[i, 'rsi'], 2),

                    'entry_rsi_past_1_week':
                        entry_data['entry_rsi_past_1_week'],

                    'entry_rsi_past_2_week':
                        entry_data['entry_rsi_past_2_week'],

                    'latest_flg':
                        'Y'
                        if i == len(stock) - 1
                        else 'N'
                }

                final_trades.append(trade)

                in_position = False

        print(f"✅ Done: {symbol}")

    except Exception as e:
        print(f"❌ Error: {symbol} -> {e}")

# ====================================
# FINAL DATAFRAME
# ====================================
result_df = pd.DataFrame(final_trades)

# ====================================
# COLUMN ORDER
# ====================================
result_df = result_df[[
    'symbol',
    'entry_date',
    'entry_price',
    'entry_rsi',
    'exit_date',
    'exit_price',
    'exit_rsi',
    'gain_pct',
    'rsi_current',
    'entry_rsi_past_1_week',
    'entry_rsi_past_2_week',
    'latest_flg'
]]

# ====================================
# SAVE TO POSTGRES
# ====================================
result_df.to_sql(
    'supertrend_weekly',
    engine,
    if_exists='replace',
    index=False
)

print("🎉 supertrend_weekly table loaded successfully")
```
