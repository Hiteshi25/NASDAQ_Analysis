```python
import yfinance as yf
import pandas as pd
import time

# 👉 load symbols from your CSV (BEST way)
import pandas as pd
df = pd.read_csv("C:\\Users\\hites\\OneDrive\\Documents\\MSC DATA SCIENCE\\PROJECTS\\NASDAQ ANALYSIS\\NASDAQ DATA\\Nasdaq_Global.csv")
symbols = df["symbol"].dropna().tolist()

# 👉 loop
for symbol in symbols:
    print(f"Downloading {symbol}...")

    try:
        data = yf.download(symbol, period="20y", interval="1d")

        if data.empty:
            print(f"⚠️ No data: {symbol}")
            continue

        data.reset_index(inplace=True)
        data["symbol"] = symbol

        data = data[[
            "symbol", "Date", "Open", "Low", "High", "Close", "Volume"
        ]]

        data.columns = [
            "symbol", "date", "open_price",
            "low_price", "high_price",
            "close_price", "volume"
        ]

        # 👉 SAVE SEPARATE FILE
        file_name = f"{symbol}_data.csv"
        data.to_csv(file_name, index=False)

        print(f"✅ Saved: {file_name}")

        time.sleep(1)   # VERY IMPORTANT

    except Exception as e:
        print(f"❌ Error: {symbol} → {e}")
```
